```python
P_yes = len(df[df['play'] == 'yes']) / len(df) # P(yes)
```

```python
P_yes
```

```python
P_no = len(df[df['play'] == 'no']) / len(df) # P(no)
```

```python
P_no
```

```python
P_cool_yes=len(df[(df['temp'] == 'cool') & (df['play'] == 'yes')]) / len(df[df['play'] == 'yes']) # P( temp=cool / yes)
```

```python
P_sunny_yes=len(df[(df['outlook'] == 'sunny') & (df['play'] == 'yes')]) / len(df[df['play'] == 'yes']) # P(outlook=sunny / yes)
```

```python
P_sunny_yes
```

```python
P_high_yes=len(df[(df['humidity'] == 'high') & (df['play'] == 'yes')]) / len(df[df['play'] == 'yes']) # P(humidity=high / yes)
```

```python
P_strong_yes=len(df[(df['windy'] == True) & (df['play'] == 'yes')]) / len(df[df['play'] == 'yes']) # P(windy=True / yes)
```

```python
P_yes_final = P_yes * P_sunny_yes * P_cool_yes * P_high_yes * P_strong_yes # P(yes) * P(sunny/yes) * P(cool/yes) * P(high/yes) * P(strong/yes)
```

```python
P_yes_final # P(yes / sunny, cool, high, strong)
```

```python
P_sunny_no = len(df[(df['outlook'] == 'sunny') & (df['play'] == 'no')]) / len(df[df['play'] == 'no']) # P(outlook=sunny / no)
```

```python
P_cool_no = len(df[(df['temp'] == 'cool') & (df['play'] == 'no')]) / len(df[df['play'] == 'no']) # P(temp=cool / no)
```

```python
P_high_no = len(df[(df['humidity'] == 'high') & (df['play'] == 'no')]) / len(df[df['play'] == 'no']) # P(humidity=high / no)
```

```python
P_strong_no = len(df[(df['windy'] == True) & (df['play'] == 'no')]) / len(df[df['play'] == 'no']) # P(windy=True / no)
```

```python
P_no_final = P_no * P_sunny_no * P_cool_no * P_high_no * P_strong_no # P(no) * P(sunny/no) * P(cool/no) * P(high/no) * P(strong/no)
```

```python
P_no_final # P(no / sunny, cool, high, strong)
```

```python
print("Yes" if P_yes_final > P_no_final else "No")
```

```python
if P_yes_final > P_no_final :
  print("Yes")
else:
  print("No")
```

USING NAIVE BAYES

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # can't use one hot encoding as assumption of having independent features will fail
```

```python
df_encoded = df.copy()
```

```python
df_encoded['outlook'] = le.fit_transform(df_encoded['outlook'])
```

```python
df_encoded['temp'] = le.fit_transform(df_encoded['temp'])
df_encoded['humidity'] = le.fit_transform(df_encoded['humidity'])
df_encoded['windy'] = le.fit_transform(df_encoded['windy'])
df_encoded['play'] = le.fit_transform(df_encoded['play'])
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
print(df_encoded['play'])
```

```python
model = CategoricalNB()
```

```python
model.fit(df_encoded.drop('play', axis=1), df_encoded['play'])
```

```python

print(le.classes_)
print
```

```python
model.predict([[2, 0,0, 1]])
```

```python
sample = pd.DataFrame([[2,0,0,1]], columns=['outlook','temp','humidity','windy'])
```

```python
prediction = model.predict(sample)
```

```python
prediction
```

```python
result = le.inverse_transform(prediction)
result
```
