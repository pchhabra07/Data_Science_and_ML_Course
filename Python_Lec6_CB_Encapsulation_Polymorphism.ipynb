## Encapsulation

> Protecting the data inside the class

> Keep data properties and method together in a class while controlling how the data can be accessed outside the class





# Private Data Member
Properties can be defined as private with attaching __ as a prefix.



> The private functionality is implemented by changing the name of the data member. The name is set as _ClassName__VariableName and can also be accessed outside using this name. This is called name mangling. This works because after name mangling, the data member with that name doesn't exist outside the class.



```python
class Person:
  def __init__(self,name,age):
    self.name=name
    self.__age=age
    print("Age:",self.__age) #Private data members can be accessed inside the member function of the class.

p1=Person("Aman",25)
print(p1.name)
print(p1._Person__age) #Ye karna nhi hai but can be done.
```

```python
class Person:
  def __init__(self,name,age):
    self.name=name
    self.__age=age
  def modifyAge(self,age): #Setter function
    self.__age=age
  def getAge(self): #Getter function
    return self.__age

p1=Person("Pratham",20)
print(p1.getAge())
p1.modifyAge(21)
print(p1.getAge())
```

# Protected Data Member
Properties can be defined as protected with attaching _ as a prefix.



> There is no name mangling in protected data members and can be accessed and modified outside the class.

> By convention, it is recommended not to be accessed or modified outside the class. Unlike private, doing so won't throw any error but isn't recommended.





```python
class Worker:
  def __init__(self,name,wage):
    self.name=name
    self._wage=wage #Protected Data Member

w1=Worker("Pratham",20000)
print(w1._wage) #Can be accessed but shouldn't be
```

## Polymorphism

> Function name remains same but they work differently depending on data types of the parameters or number of parameters.

> Eg. len("Pratham") gives the length of the string and counts the number of characters.

> len([1,2,3]) counts the number of elements in the list and gives the length.

> Hence, the functionality of the len() function changed with the data type of the parameter.









Allows different classes to define methods with the same method name and python decide which method to run based on the object.

```python
class Car:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model
  def move(self):
    print("Drive")
class Boat:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model
  def move(self):
    print("Sail")
class Airplane:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model
  def move(self):
    print("Fly")

car1=Car("Tesla","Model S")
boat1=Boat("Yamaha","Waverunner")
airplane1=Airplane("Airbus","A320")

for x in (car1,boat1,airplane1):
  x.move()
```

```python
class Vehicle:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model

class Car(Vehicle):
  def move(self):
    print(f"{self.model} of {self.brand} is driving.")

class Boat(Vehicle):
  def move(self):
    print(f"{self.model} of {self.brand} is sailing.")

class Airplane(Vehicle):
  def move(self):
    print(f"{self.model} of {self.brand} is flying.")

car1=Car("Tesla","Model S")
boat1=Boat("Yamaha","Waverunner")
airplane1=Airplane("Airbus","A320")

for x in (car1,boat1,airplane1):
  x.move()
```
