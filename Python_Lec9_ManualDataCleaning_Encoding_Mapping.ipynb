# Rules for missing data

> Either fill the missing values of the column (done when number of missing values in a column is very low, say 10%)

> Or Drop the whole column which has missing values (done when number of missing values in a column is large, say more than 30-40%)

## Encoding

> When a column has more than 2 types of values (unlike boolean or male female etc.), we can't map them to differnet numerical values.

> When we represent a column as 0 and 1, its representation.

> When we represent a column as 0, 1 and 2 or more, it defines priority rather than representation.

> Hence, we don't map values in case of more than 2 unique values (we do when there is actual priority in the unique values. Eg. pclass column of this dataset gives passenger class as first class, second class or third class with first class being top priority). We perform encoding.









We perform One Hot Encoding. Lets take example for 'embarked' column,

*   In 'embarked' column, three categories of values are present. 'C', 'S' and 'Q'.
*   get_dummies() function of pandas library takes a list of columns we need to apply this on.

*   It drops the original column.
*   It makes new columns or dummies, one for each unique value (here 'embarked_C', 'embarked_S' and 'embarked_Q') in alphabetical order.

*   For value 'C' in embarked, the value for 'embarked_C' will be 1 or True and for rest will be 0 or False and so on.







```python
#df=pd.get_dummies(df, columns=['embarked'])
```

```python
#df.head()
```

We noticed that we can reduce number of columns by 1 to this as well.

For that we use 'drop_first=True' attribute which removes the first dummy according to alphabetical order (here 'embarked_C').

False and False value for 'embarked_S' and 'embarked_Q' respectively represents the value 'C'.

```python
df=pd.get_dummies(df, columns=['embarked'],drop_first=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df, columns=['who'],drop_first=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df, columns=['embark_town'],drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

For class column, we will use mapping as we have to set top priority to first class, then second and then third.
