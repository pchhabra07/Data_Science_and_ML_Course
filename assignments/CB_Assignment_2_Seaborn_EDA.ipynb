## B. Numerical Columns

```python
sns.histplot(df['age'], bins=10, kde=True)

# Data is not normally distributed.
# Data is Right Skewed or Positively Skewed (i.e. flat longer tail on the right side denoting outliers)
# Data peaks at 20 to 40 years of age.
```

```python
sns.histplot(df['fare'], bins=10, kde=True)

# This is a heavily Right Skewed or Positively Skewed Graph.
# This is not normally distributed as well.
# Most people paid $30-$40 for their fare
```

```python
sns.boxplot(df['age'], orient='h')

# Minimum   Age: 0 years old
# 25th %ile Age: 20 years old (i.e. 25% people are of age 20 years old or less)
# 50th %ile Age: 28 years old (i.e. 50% people are of the age 28 years or less, 25% people are of 20-28 years old)
# 75th %ile Age: 38 years old (i.e. 75% people are of the age 38 years or less, 25% people are of 28-38 years old)
# Maximum   Age: 65 years old
# Few Outliers of age 65 to 80 years old
```

```python
sns.boxplot(df['fare'], orient='h')

# Minimum Fare: Around $5
# 25%ile  Fare: $10 (i.e. 25% people paid $5 to $10 for their fare)
# 50%ile  Fare: $20 (i.e. 25% people paid $10 to $20 for their fare)
# 75%ile  Fare: $45 (i.e. 25% people paid $20 to $45 for their fare)
# Maximum Fare: $75 (i.e. and the rest 25% paid $45 to $75 for their fare)
# Few outliers from $80 to $500 (i.e. only few compared to the rest paid a hefty amount of $80 to $500)
```

# Part 3 (Multivariate Analysis):-

## A. Numerical vs Numerical Analysis

```python
sns.scatterplot(df,x='age',y='fare')

# Data shows that irrespective of the age of a passenger,
# most of the passengers paid under $100 for their fare.
# There isn't much correlation between Age and Fare as Majority paid a similar fair,
# except for a few outliers.
```

## B. Numerical vs Categorical Analysis

```python
sns.barplot(df,x='sex',y='age')

# This shows that avg. Age of Males is around 31 years whereas avg. Age of Females is around 26-27 years.
# Average Age of Males is slightly more than Females.
```

```python
sns.barplot(df, x='survived', y='fare')

# This clearly shows the avg. Fare of passengers who survived was more than half
# of the avg. Fare of passengers who did not survive.

# Although, Survivors show some error as well as the error line is comparitively
# longer than those who didn't survive. Hence, the estimate of avg. Fare of
# survivors is not as precise as the estimate of avg. Fare of victims.

# It depicts variability in data of survivors.
```

```python
sns.barplot(df, x='pclass', y='fare')

# This shows that the avg. Fare of First passenger class was 4 times more
# than that of the avg. Fare of Second and Third passenger class travellers.

# The estimate of avg. Fare of First passenger class travellers was also less
# precise as compared to Second and Third passenger class travellers.

# It depicts variability in data of passenger class.
```

## C. Categorical vs Categorical Analysis

```python
sns.heatmap(pd.crosstab(df['sex'],df['survived']))

# This depicts that about 100 to 125 Male travellers managed to survive,
# and about 450 Male travellers couldn't survive, whereas about 225 to 250 Females
# managed to survive and less that 100 Females couldn't survive.

# This depicts females were prioritised for survival.
```

```python
sns.heatmap(pd.crosstab(df['pclass'],df['survived']))

# This depicts that 350 or more Third class passengers did not survive.
# About 125 of Third class passengers managed to survive.
# The survival rate of Second class passengers also wasn't good but for First class
# passengers, more than 150 survived and less than 100 did not survive.

# This depicts that First class passengers were prioritised more.
```

## 10 key insights from the Titanic dataset (Conclusion):

1.   Females had a significantly higher survival rate than males.

2.   Passenger class was a major factor. First Class passengers had the highest survival rate, while Third Class had the lowest.

3.   Out of the victims, the vast majority were male, whereas more than two-thirds of the survivors were female.

4.   The average fare for First Class was more than four times higher than that of Second or Third Class.

5.   Survivors, on average, paid a significantly higher fare than those who did not survive.

6.   The majority of passengers were between 20 and 40 years old, with the data being positively skewed toward younger individuals.

7.   Southampton was the most frequent embarkation town.

8.   Over 350 Third Class passengers did not survive, which is the highest casualty count among all classes.

9.   The average age of male passengers (31 years) was slightly higher than that of female passengers (26-27 years).

10.  The high variability (longer error bars) in the fares of survivors in the 'fare' vs 'survived' bar plot.
