## Linear Regression

*   This is Manual Implementation of Single Linear Regression.
*   Refer to notes for derivation of formulas for m (optimal weightage) and b (optimal offset).



```python
class LINEARREGRESSION:

    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):

        # calculate means
        x_mean = X_train.mean()
        y_mean = y_train.mean()

        # vectorized numerator
        num = np.sum((X_train - x_mean) * (y_train - y_mean))

        # vectorized denominator
        den = np.sum((X_train - x_mean) ** 2)

        # slope and intercept
        self.m = num / den
        self.b = y_mean - (self.m * x_mean)

        print("Slope (m):", self.m)
        print("Intercept (b):", self.b)


    def predict(self, X_test):

        return self.m * X_test + self.b
```

```python

df = pd.read_csv("placement.csv")

X = df.iloc[:,0].values # All rows of 0th column
y = df.iloc[:,1].values # All rows of 1st column
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2
) # random_state=2 as data is small
```

```python
X_test
```

```python
lr = LINEARREGRESSION()
lr.fit(X_train, y_train)
```

```python

print("Input CGPA:", X_test[0])
print("Predicted Package:", lr.predict(X_test[0]))
```

## Linear Regression

*   This is SKlearn Implementation of Single Linear Regression.
*   Optimal Weightage or slope or 'm' is called coefficient in sk-learn.



```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("placement.csv")

# Input and output
X = df.iloc[:,0:1]
y = df.iloc[:,-1]

# Create model
lr = LinearRegression()

# Train model
lr.fit(X, y)

# Get slope and intercept
m = lr.coef_[0]
b = lr.intercept_

print("Slope (m):", m)
print("Intercept (b):", b)

# Using equation y = mx + b
cgpa = 6.7
predicted_package = m * cgpa + b

print("Predicted package for CGPA", cgpa, ":", predicted_package)
```

## Key Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# reshape X_test to 2D
# reshape X_test correctly
X_test = X_test.reshape(-1,1)

# prediction
y_pred = lr.predict(X_test)

print("Predictions:", y_pred[:5])
print("Actual:", y_test[:5])

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)

mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

## Multiple Linear Regression


*   This is the Manual Implementation of Multiple Linear Regression.
*   



```python
import numpy as np

class LinearRegressionCustom:

    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X_train, y_train):
        # add column of 1s for intercept
        X_train = np.insert(X_train, 0, 1, axis=1)

        # calculate coefficients
        be = np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)

        self.intercept_ = be[0]
        self.coef_ = be[1:]

    def predict(self, X_test):
        y_pred = np.dot(X_test, self.coef_) + self.intercept_
        return y_pred
```
