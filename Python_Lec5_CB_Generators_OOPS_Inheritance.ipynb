## Generators

> Generators are functions that can pause and resume the execution.

> The execution starts and pauses after each "yield" statement and yield works like "return" and returns a value each time





```python
def my_generator():
  yield 3
  print('first yield done')
  yield 4
  print('second yield done')
  yield 5
  print('third yield done')

for value in my_generator():
  print(value)
```



> Yield pauses the program execution and is used to return multiple values while "return" ends the function's execution all at once and is used to return a single value.



```python
def my_generator():
  yield 3
  yield 4
  yield 5

gen=my_generator()
print(gen)
```

```python
def count_up_to(n):
  c=1
  while c<=n:
    yield c
    c=c+1

for num in count_up_to(4):
  print(num)
```

```python
def largeSequence(n):
  for i in range(n):
    yield i

gen=largeSequence(200000000)
print(next(gen))
print(next(gen))
```

```python
def normal():
  return 1
def generator():
  yield 1
print(type(normal()))
print(type(generator()))
```

## init

> Each class has an init method and it always runs first upon object creation and is used to initialise the object.

> It is a built in method by default but can be user defined.

> It is denoted as "_ _ init _ _"

```python
class Person:
  def __init__(self, name, age):
    self.name=name
    self.age=age

p1=Person('Pratham',20)
print(p1.name)
print(p1.age)
```

## Class Properties

> Variables that belong to the class are called as Class properties.

> These variables are same for all objects of that class.

## Object Properties

> Variables that belong to an object only and not the whole class are called as Object Properties.

> These variables are different for all objects of that class.

```python
class Person:
  gender="Male" #Class Properties
  def __init__(self, name):
    self.name=name #Object Properties

p1=Person('Pratham')
p2=Person('Rohan')
print(p1.name)
print(p1.gender)
print(p2.name)
print(p2.gender)
```

If we want to update a Class Property, we can update it directly by writing "class_name.class_property_name=updated_value"

```python
class Person:
  gender="Male" #Class Properties
  def __init__(self, name):
    self.name=name #Object Properties

p1=Person('Pratham')
p2=Person('Rohan')
print(p1.name)
print(p1.gender)
print(p2.name)
print(p2.gender)

Person.gender="Female" #Updating Class Property

print(p1.name)
print(p1.gender)
print(p2.name)
print(p2.gender)
```

```python
class Person:
  gender="Male" #Class Properties
  def __init__(self, name):
    self.name=name #Object Properties

p1=Person('Pratham')
p2=Person('Rohan')

print(p1.name,p1.gender)
print(p2.name, p2.gender)

p2.gender='Bisexual'#Updating the Class Property of just one object
print(p2.name, p2.gender)

#Updating Class Property for whole class (won't change for
#those objects whose Class Property was updated, for those objects, the class property became an object property)
Person.gender="Female"

print(p1.name,p1.gender)
print(p2.name, p2.gender)
```

## Inheritance

> When a class inherits methods and properties from another class, its called Inheritance.

> Parent class or Base class is the class from which, another class inherits properties and methods.

> Child class or Derived class is the class that inherits properties and methods from another class.







```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  pass

d=Dog()
d.speak()
```

```python
class Father:
  def skill1(self):
    print("Father: Driving")

class Mother:
  def skill2(self):
    print("Mother: Cooking")

class Child(Father,Mother):
  pass

c=Child()
c.skill1()
c.skill2()
```

```python
class Person:
  def __init__(self, fname, lname):
    self.firstName=fname
    self.lastName=lname
  def printName(self):
    print(self.firstName," ",self.lastName)

class Student(Person):
  def __init__(self, fname, lname):
    pass

s=Student("Aman","Gupta")
s.printName() #This will give error as Student class has its own init method,
#so it will not use the init method of its parent and will override it.
```

```python
class Person:
  def __init__(self, fname, lname):
    self.firstName=fname
    self.lastName=lname
  def printName(self):
    print(self.firstName," ",self.lastName)

class Student(Person):
  def __init__(self, fname, lname):
    Person.__init__(self,fname,lname) #Even after overriding the __init__, we can call the Parent class's init

s=Student("Aman","Gupta")
s.printName()
```

## Super

> Function used to refer to the parent class.

> No need to use the name of the parent class as above

```python
class Person:
  def __init__(self, fname, lname):
    self.firstName=fname
    self.lastName=lname
  def printName(self):
    print(self.firstName," ",self.lastName)

class Student(Person):
  def __init__(self, fname, lname):
    #We can also refer to the parent class using "super()"
    super().__init__(fname,lname) #Person.__init__(self,fname,lname)

s=Student("Aman","Gupta")
s.printName()
```
