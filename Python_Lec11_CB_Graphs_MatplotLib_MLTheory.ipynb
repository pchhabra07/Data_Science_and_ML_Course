## Graphs

> Used for visualizing data

> Matplotlib library's 'pyplot' is used.





```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_csv('Titanic_cleaned.csv')
```

To make a histogram, we use .hist() function and we can define the number of bins using the bin parameter.

```python
plt.hist(df['age'],bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

##Boolean Indexing / Filetring:-

*   df['survived']==1 #This returns all the rows in the dataframe and assigns "True" to those with 'survived'=1 and False to others.
*   df[df['survived']==1]['age'] #This fetches the age column but only shows those rows whose 'survived'=1 or basically value is True for "df['survived']==1".



```python
df['survived']==1
```

```python
df[df['survived']==1]['age']
```

```python
plt.hist(df[df['survived']==1]['age'],bins=20)
plt.xlabel('Age of Survivors')
plt.ylabel('Count of Survivors')
plt.title('Age Distribution of Survived')
plt.show()
```

```python
plt.hist(df[df['survived']==1]['age'],bins=20,alpha=0.5,label='Survived') #alpha controls the opacity of the histogram bars.
plt.hist(df[df['survived']==0]['age'],bins=20,alpha=0.5,label='Not Survived') #it enables us to view overlapping histogram graphs.
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend() #Gives color for categories, takes from "label" parameter from .hist() command
plt.title('Age Distribution of Survived Vs Not Survived')
plt.show()
```

```python
df[df['survived']==1]['fare']
```

```python
plt.hist(df[df['survived']==1]['fare'],bins=20,alpha=0.5,label='Survived')
plt.hist(df[df['survived']==0]['fare'],bins=20,alpha=0.5,label='Not Survived')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Fare Distribution of Survivors Vs Non-Survivors')
plt.legend()
plt.show()
```

We see that the column 'sibsp' (number of sibling and spouse who came to the ship along) and the column 'parch' (parents and children who came to the ship along) don't make sense being separate. We can make a single column "family_size" which has sum of both 'sibsp' and 'parch' and +1 (themselves).

```python
df['family_size']=df['parch']+df['sibsp']+1
```

```python
df.groupby('family_size')['survived'].mean()
```

## Machine Learning

*   Traditional Programming takes input and logic/rule and the machine uses that to generate output.

*   Ex. We give a machine a=5, b=4 and sum=a+b, machine returns 9.
*   Machine Learning takes input and the output and takes numerous such data and learns from it and gives what the logic/rule is used in the data through patterns.

*   Ex. We give a machine a=3,b=2 and sum=5, and many such examples. The machine returns the logic/rule after learning from the data as sum=a+b.

## ML Vs DL

*   Machine Learning takes input values (features) and output values(target value) manually and recognises patterns.

*   Ex. we provide features like "CGPA", "IQ Level" and target values like "Placement" and recognise the patter amongst those columns.
*   Deep Learning extracts features automatically through layers.


*   Ex. We provide images of Dogs and Cats and target value that which is Dog and which is Cat and the model automatically extracts features like "Long Nose" -> Dog, "Moustaches and Blue Eyes" -> Cat

## Types of Machine Learning

1.   Supervised Learning:

    *   Definite features and target value are given.
    *   Implemented through a) Regression (numerical data)
                            b) Classification (categorical data)


2.   Unsupervised Learning:


    *   Only input values or features are given and no target value is given
    *   Implemented through Clustering or clustering similar type of data in a cluster.


3.   Semi Supervised Learning:

    *   70% data is labelled and 30% data is unlabelled and the model automatically labels the data.

4.   Reinforcement Learning:

    *   When our model learns from trial and error and avoids future mistakes.







Two differents ways of training a model:-

1.   Batch Learning (offline learning):

    *   Entire dataset at once for the training of the model.
    *   Collect all training data and train the model on the server
    *   Slow for huge datasets
    *   Cannot adapt quickly to the new dataset.
    *   Used for dataset that is static and not dynamic (does not change in the future) like Historical data.

2.   Online Learning (model learning):

    *   Model learns continuously from the incoming data.
    *   Data comes one by one.
    *   Model updates itself continuously.
    *   Ex. Stock Price Prediction
