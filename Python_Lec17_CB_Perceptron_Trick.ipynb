## Moving the separable line by only considering Misclassified Points:-

```python
import numpy as np

# 1. Small simple dataset
X = np.array([[1, 1],
              [2, 1],
              [4, 3],
              [5, 4]])

y = np.array([0, 0, 1, 1])   # 0 = negative, 1 = positive

# 2. Initialize
w = np.array([0.0, 0.0])  # weights
b = 0.0                   # bias
lr = 1                    # learning rate (keep 1 for simplicity)

# 3. Training (just 1 loop to understand)
for i in range(len(X)):

    x = X[i]
    true = y[i]

    # prediction
    z = w[0]*x[0] + w[1]*x[1] + b

    if z >= 0:
        pred = 1
    else:
        pred = 0

    print(f"\nPoint: {x}, True: {true}, Pred: {pred}")

    # 4. Perceptron Trick
    if pred != true:
        print(" Wrong → Update")

        if true == 1:
            w = w + lr * x
            b = b + lr
        else:
            w = w - lr * x
            b = b - lr

    else:
        print(" Correct → No update")

    print("Weights:", w, "Bias:", b)
```

## Moving the separable line by considering Classified & Misclassified Points (using Sigmoid Activation Function):-

```python
import numpy as np

# 1. Dataset (same as before)
X = np.array([[1, 1],
              [2, 1],
              [4, 3],
              [5, 4]])

y = np.array([0, 0, 1, 1])

# 2. Initialize
w = np.array([0.0, 0.0])
b = 0.0
lr = 0.1

# 3. Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 4. Training
for epoch in range(5):   # multiple passes

    print(f"\n--- Epoch {epoch+1} ---")

    for i in range(len(X)):

        x = X[i]
        true = y[i]

        # Step 1: Linear equation
        z = w[0]*x[0] + w[1]*x[1] + b

        # Step 2: Probability
        pred = sigmoid(z)

        # Step 3: Error
        error = true - pred

        print(f"x={x}, true={true}, pred={round(pred,3)}, error={round(error,3)}")

        # Step 4: Update (Gradient Descent)
        w = w + lr * error * x
        b = b + lr * error

        print(f"Updated w={w}, b={b}")
```

```python
import numpy as np

# Same dataset
X = np.array([[1, 1],
              [2, 1],
              [4, 3],
              [5, 4]])

y = np.array([0, 0, 1, 1])
```

```python


def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def perceptron(X, y):

    # Add bias column
    X = np.insert(X, 0, 1, axis=1)

    # Initialize weights
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(1000):

        j = np.random.randint(0, len(X))

        y_hat = sigmoid(np.dot(X[j], weights))

        weights = weights + lr * (y[j] - y_hat) * X[j]

    return weights[0], weights[1:]


#  CALL FUNCTION
b, w = perceptron(X, y)

print("Bias:", b)
print("Weights:", w)
```
