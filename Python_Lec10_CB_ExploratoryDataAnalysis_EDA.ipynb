```python
import pandas as pd
df=pd.read_csv('Titanic_cleaned.csv')
df.shape
```

To count values of a column grouping by category (0 means 'no' and 1 means 'yes')

```python
df['survived'].value_counts()
```

.describe() function describes the whole dataframe by giving the count, mean, standard deviation, min, 25%ile, 50%ile, 75%ile, max.

```python
df.describe()
```

To find how many males survived and how many females survived and how many males didn't survive and how many females didn't survive, we need to find the relation between the two columns 'survived' and 'sex'.

We use .crosstab() function with both columns as arguments. (works only for both columns being categorical data)

```python
pd.crosstab(df['sex'],df['survived'])
```

If we need the percentage of how many females survived and how many males survived, etc.

We need to normalize by index (divide the value by the sum of values in the row)

```python
pd.crosstab(df['sex'],df['survived'],normalize='index')
```

If we need the percentage of how many people who survived were females and how many people who survived were males, etc.

We need to normalize by column (divide the value by the sum of values in the column)

```python
pd.crosstab(df['sex'],df['survived'],normalize='columns')
```

Find relation between class and survived. class 1 means Third Class and class 3 means First class

```python
pd.crosstab(df['class'],df['survived'])
```

```python
pd.crosstab(df['class'],df['survived'],normalize='index')
```

For non-categorical columns or continuous data like 'age', we can't use .crosstab() as it will take all unique values of the column and categorise it based on that.

```python
pd.crosstab(df['age'],df['survived'])
```

For continuous data, we use column.groupby(grouping_column). We select the column and apply a groupby on it based on the categorical column.

Hence, we use this when one column is continuous and one column is categorical and we need an aggregate value of the continuous column.

```python
df['age'].groupby(df['survived']).mean()
```

```python
df.groupby('survived')['fare'].mean()
```
