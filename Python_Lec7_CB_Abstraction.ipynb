## Abstraction

> hiding internal details (implementation) and showing important features only.

> Parent class doesn't show how the abstracted method is working.

> Also, a rule is enforced that all child classes of the Parent Abstract Class must have their own implementation of the Abstract Method, else we get an error.







> Eg. A parent class or base class or Abstract Class is MainBankApp. It has an abstract method with no implementation. The child classes MobileBankApp and WebBankApp inherit MainBankApp.

> We cannot make an object of the Abstract Class.

> The child class must have the implementation of the Abstract Method. If there is no implementation, we get an error.



```python
from abc import ABC,abstractmethod

class MainBankApp(ABC):
  @abstractmethod
  def security(self):
    pass

class MobileBankApp(MainBankApp):
  def security(self):
    print("Security Method")

  def mobile_log_in(self):
    print('login into mobile')

class WebBankApp(MainBankApp):
  def web_log_in(self):
    print('login into web')

mob=MobileBankApp()
mob.security()

# The following line will now correctly give an error as WebBankApp doesn't implement the abstract method
#web=WebBankApp()

# The following line will now correctly give an error as object of Abstract Class can't be made
# as the abstract method has no implementation there due to Abstraction
#main=MainBankApp()
```
