```python
#Question 15
n=int(input("Enter size:"))
lst=[]
for i in range(n):
  x=int(input("Enter numbers:"))
  lst.append(x)

mx=lst[0]
mxIdx=0
for i in range(1,n):
  if lst[i]>mx:
    mx=lst[i]
    mxIdx=i

if mxIdx==0:
  mx2=lst[1]
  for i in range(2,n):
    if lst[i]>mx2:
      mx2=lst[i]
  print(mx2)
else:
  mx2=lst[0]
  for i in range(1,n):
    if i==mxIdx:
      continue
    if lst[i]>mx2:
      mx2=lst[i]
  print(mx2)
```

```python
#Question 16
n=int(input("Enter size:"))
lst=[]
visited={}
for i in range(n):
  x=int(input("Enter numbers:"))
  lst.append(x)
  visited[x]=False

withoutDuplicates=[]
for i in lst:
  if visited[i]==True:
    continue
  else:
    withoutDuplicates.append(i)
    visited[i]=True
print(withoutDuplicates)
```

```python
#Question 17
s1=input("Enter string 1:")
s2=input("Enter string 2:")

s1dict={}
s2dict={}

for i in s1:
  if i in s1dict.keys():
    s1dict[i]+=1
  else:
    s1dict[i]=1

for i in s2:
  if i in s2dict.keys():
    s2dict[i]+=1
  else:
    s2dict[i]=1

if s1dict==s2dict:
  print("Anagram")
else:
  print("Not Anagram")
```

```python
#Question 19
def swapWithTemp(a,b):
  temp=a
  a=b
  b=temp
  return a,b
a=10
b=5
a,b=swapWithTemp(a,b)
print(a,b)

def swapWithoutTemp(a,b):
  a=a+b
  b=a-b
  a=a-b
  return a,b
a=10
b=5
a,b=swapWithoutTemp(a,b)
print(a,b)
```

```python
#Question 20
def isPalindrome(myTuple):
  i=0
  j=len(myTuple)-1
  while i<j:
    if myTuple[i]!=myTuple[j]:
      return "Not Palindrome"
    i+=1
    j-=1
  return "Palindrome"
myTuple1=(1,2,3,2,1)
print(isPalindrome(myTuple1))
myTuple2=(1,2,3,4,5)
print(isPalindrome(myTuple2))
```
