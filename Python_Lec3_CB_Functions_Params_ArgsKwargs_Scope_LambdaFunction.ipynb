## Parameters and Arguments

> Parameters are the variables that are used in the function on definition.

> Arguments are the values used to call a function.





```python
def sumFunc(a,b):
  #a and b are parameters
  return a+b

#2 and 6 are arguments
print(sumFunc(2,6))
```

```python
def my_func(name,lastname):
  print(name+lastname)

#This function call is invalid as it sends only one argument where 2 are required
my_func("Aryesh")
```

## Args and Kwargs

> Args means Arbitrary arguments

> Kwargs means arbitrary keyword arguments





```python
def my_func(name,lname):
  print(name+lname)

#Positional Arguments as Aman goes to name and Gupta goes to lname
my_func("Aman","Gupta")

#Keyword Arguments as we are defining which value to go in which parameter
my_func(lname="Ayush",name="Sharma")
```

## Positional Only Arguments

> When a slash is used in "def func_name(a,b,/):", it means the parameters before the / can only accept positional arguments and not keyword arguments



```python
def my_fun(name,/):
  print("Hello ",name)
my_fun("Aman")
my_fun(name="Aman") #Error as keyword arguments not allowed
```

## Keyword Only Arguments

> When a star is used in "def func_name(*,a,b):", it means the parameters after the * can only accept keyword arguments and not positional arguments.



```python
def my_fun(*,name):
  print("Hello ",name)
my_fun("Aman") #Error as positional arguments not allowed
my_fun(name="Aman")
```

## *args and **kwargs

> args is just a name for arguments and kwargs is just a name for keyword arguments

> *args allow to accept any number of positional arguments and forms a tuple out of it. " * " is important for this.





```python
def my_func(*name):
  print("my name is",name[0])
my_func("Aman","Rahul","Akshay")
```

```python
def my_func(x,*y):
  print(x,y[2])
my_func("Ayush","Aman","Rahul","Akshay")
```



> **kwargs is used to accept any number of keyword arguments and forms a dictionary out of it. " ** " is important for this



```python
def my_func(**x):
  print("My name is",x['name'])
my_func(name='Virat',lname='Kohli',team='RCB')
```

```python
def my_function(a,b,c):
  return a+b+c

numbers=[1,2,3]
result=my_function(*numbers) #Same as my_function(1,2,3)
# * is used to unpack
print(result)
```

```python
def my_function(name,age):
  print(name,age)

person={'name':'Pratham','age':20}
my_function(**person) #Same as my_function(name='Pratham',age=20)
# ** is used to unpack keyword arguments from a dictionary
```

```python
def my_function():
  x=300
  def inner_func():
    print(x)
  inner_func()
my_function()
```

## Local and Global Scope

```python
#Global Variable
x=300
def myfunc():
  #Local Variable
  x=200
  print(x) #prints the local variable

myfunc()
#prints the global variable as the global wasn't changed
print(x)
```



> To change a global variable inside a function body, we use the "global" keyword to denote a variable as the global variable.



```python
#Global Variable
x=300
def myfunc():
  #global variable
  global x
  x=200
  print(x) #prints the global updated variable

myfunc()
#prints the global variable as the global was changed
print(x)
```

## Lambda Function

> small function written in one line

> used for short simple operation

# Properties:-

> no function name

> no return keyword

> can take any number of arguments

> can have only one exception

> automatically returns the result

```python
y = lambda x : x + 10
print(y(5))
```

```python
y = lambda a , b : a * b
print(y(5,10))
```

```python
def myfunc(n):
  return lambda a:a*n

mydoubler=myfunc(2)
mytripler=myfunc(3)
print(mydoubler)
print(mydoubler(11))
print(mytripler(11))
```

```python
numbers = [1,2,3,4,5]
# map() is a function that takes a lambda function and iterates through the
#whole list and applies that lambda function to each element of the list
doubled = list(map(lambda x:x*2,numbers))

print(doubled)
```

```python
numbers = [1,2,3,4,5]
# filter() is a function that takes a lambda function and iterates through the whole list and takes a boolean value,
#if True, keeps that element in that list, else filters it out.
oddNums = list(filter(lambda x:x%2!=0,numbers))

print(oddNums)
```
