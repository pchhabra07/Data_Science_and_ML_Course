# **Batch Gradient Descent (BGD)**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1 : Create simple dataset


# X = study hours
X = np.array([1,2,3,4,5])

# Y = exam marks
y = np.array([3,5,7,9,11])
```

```python
 # Step 2 : Initialize parameters


m = 0   # slope
b = 0   # intercept

learning_rate = 0.01
epochs = 10

n = len(X) # number of rows

print("Initial m:", m)
print("Initial b:", b)
```

```python
# Step 3 : Gradient Descent


for i in range(epochs):

    # the formula 'm*X + b' is applied to each element of 'X' array and a new array is made called 'y_pred'
    y_pred = m*X + b

    # calculate gradients
    dm = (-2/n) * sum(X * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)

    # update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # calculate cost
    cost = (1/n) * sum((y - y_pred)**2)

    print("Epoch:",i+1)
    print("m =",round(m,3)," b =",round(b,3)," cost =",round(cost,3))
```

```python
#Step 4 : Final model


print("Final Model:")
print("y =",round(m,2),"x +",round(b,2))
```

```python
# Step 5 : Plot result


plt.scatter(X,y,color="blue",label="Data") # represents actual data points

y_final = m*X + b

plt.plot(X,y_final,color="red",label="Regression Line") # represents regression model line after 10 epochs

plt.legend()
plt.show()
```

# **Stochastic Gradient Descent**


```python
import numpy as np
import matplotlib.pyplot as plt


# Step 1 : Dataset

X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])

# Step 2 : Initialize parameters

m = 0
b = 0

learning_rate = 0.01
epochs = 15

n = len(X)

# Step 3 : Stochastic Gradient Descent

for epoch in range(epochs):

    for i in range(n):
        x_i = X[i]
        y_i = y[i]

        # predicted value
        y_pred = m*x_i + b

        # gradients
        dm = -2 * x_i * (y_i - y_pred)
        db = -2 * (y_i - y_pred)

        # update parameters
        m = m - learning_rate * dm
        b = b - learning_rate * db

    # calculate cost
    y_pred_all = m*X + b
    cost = np.mean((y - y_pred_all)**2)

    print("Epoch:",epoch+1)
    print("m =",round(m,3)," b =",round(b,3)," cost =",round(cost,3))

print("Final Model:")
print("y =",round(m,2),"x +",round(b,2))
```

# **Mini-Batch Gradient Descent**

```python
import numpy as np

# dataset
X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])

m = 0
b = 0

learning_rate = 0.01
epochs = 5
batch_size = 2

n = len(X)

for epoch in range(epochs):

    for i in range(0,n,batch_size):

        X_batch = X[i:i+batch_size]
        y_batch = y[i:i+batch_size]

        y_pred = m*X_batch + b

        dm = (-2/len(X_batch)) * np.sum(X_batch*(y_batch - y_pred))
        db = (-2/len(X_batch)) * np.sum(y_batch - y_pred)

        m = m - learning_rate * dm
        b = b - learning_rate * db

    # calculate cost
    y_pred_all = m*X + b
    cost = np.mean((y - y_pred_all)**2)

    print("Epoch:",epoch+1," m:",round(m,3)," b:",round(b,3), " cost:",round(cost,3))

print("Final Model:")
print("y =",round(m,2),"x +",round(b,2))
```

# **Plot for SGD**
                          

| Epoch 1 | model after first pass through data  
| Epoch 2 | model improved                      
| Epoch 3 | getting closer                      
| Epoch 4 | almost correct                      
| Epoch 5 | near best line                      


```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset

X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])


# Initialize parameters

m = 0
b = 0

learning_rate = 0.01
epochs = 5

n = len(X)


m_history = []
b_history = []

# SGD Training

for epoch in range(epochs):

    for i in range(n):

        x_i = X[i]
        y_i = y[i]

        y_pred = m*x_i + b

        # gradients
        dm = -2 * x_i * (y_i - y_pred)
        db = -2 * (y_i - y_pred)

        # update
        m = m - learning_rate * dm
        b = b - learning_rate * db

    # store parameters after epoch
    m_history.append(m)
    b_history.append(b)


# Plot Data

plt.scatter(X,y,color="blue",label="Data")

# plot regression line for each epoch
for i in range(len(m_history)):

    y_line = m_history[i]*X + b_history[i]

    plt.plot(X,y_line,label=f"Epoch {i+1}")

plt.legend()
plt.title("SGD Learning Progress")
plt.show()
```

# **Plot for Mini-Batch Gradient Descent**

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1 : Dataset

X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])


# Step 2 : Initialize parameters

m = 0
b = 0

learning_rate = 0.01
epochs = 5
batch_size = 2

n = len(X)

# store history for plotting
m_history = []
b_history = []


# Step 3 : Mini Batch Gradient Descent

for epoch in range(epochs):

    for i in range(0, n, batch_size):

        # create batch
        X_batch = X[i:i+batch_size]
        y_batch = y[i:i+batch_size]

        # predictions
        y_pred = m*X_batch + b

        # gradients
        dm = (-2/len(X_batch)) * np.sum(X_batch * (y_batch - y_pred))
        db = (-2/len(X_batch)) * np.sum(y_batch - y_pred)

        # update parameters
        m = m - learning_rate * dm
        b = b - learning_rate * db

    # store line after each epoch
    m_history.append(m)
    b_history.append(b)


# Step 4 : Plot data

plt.scatter(X, y, color="blue", label="Data")

# plot line for each epoch
for i in range(len(m_history)):

    y_line = m_history[i]*X + b_history[i]

    plt.plot(X, y_line, label=f"Epoch {i+1}")

plt.title("Mini Batch Gradient Descent Learning Progress")
plt.legend()
plt.show()
```

# **Comparison Plot of all three**

```python
import numpy as np
import matplotlib.pyplot as plt


# Dataset

X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])

learning_rate = 0.01
epochs = 5
n = len(X)


# Batch Gradient Descent

m_b, b_b = 0,0
batch_history = []

for epoch in range(epochs):

    y_pred = m_b*X + b_b

    dm = (-2/n) * np.sum(X*(y-y_pred))
    db = (-2/n) * np.sum(y-y_pred)

    m_b = m_b - learning_rate*dm
    b_b = b_b - learning_rate*db

    batch_history.append((m_b,b_b))

# Stochastic Gradient Descent

m_s, b_s = 0,0
sgd_history = []

for epoch in range(epochs):

    for i in range(n):

        y_pred = m_s*X[i] + b_s

        dm = -2*X[i]*(y[i]-y_pred)
        db = -2*(y[i]-y_pred)

        m_s = m_s - learning_rate*dm
        b_s = b_s - learning_rate*db

    sgd_history.append((m_s,b_s))


# Mini Batch Gradient Descent

batch_size = 2
m_m, b_m = 0,0
mini_history = []

for epoch in range(epochs):

    for i in range(0,n,batch_size):

        X_batch = X[i:i+batch_size]
        y_batch = y[i:i+batch_size]

        y_pred = m_m*X_batch + b_m

        dm = (-2/len(X_batch))*np.sum(X_batch*(y_batch-y_pred))
        db = (-2/len(X_batch))*np.sum(y_batch-y_pred)

        m_m = m_m - learning_rate*dm
        b_m = b_m - learning_rate*db

    mini_history.append((m_m,b_m))



# Plotting

fig, ax = plt.subplots(1,3,figsize=(15,4))

titles = ["Batch Gradient Descent","Stochastic Gradient Descent","Mini Batch GD"]
histories = [batch_history, sgd_history, mini_history]

for k in range(3):

    ax[k].scatter(X,y,color="blue")

    for i,(m,b) in enumerate(histories[k]):

        y_line = m*X + b
        ax[k].plot(X,y_line,label=f"Epoch {i+1}")

    ax[k].set_title(titles[k])
    ax[k].legend()

plt.show()
```
