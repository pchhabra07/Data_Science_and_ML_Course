## 2. Scaling

**MinMax Scaler**

The most intuitive way is as follows:-

*   Lets say 'X' is [1,2,3,4,5,6] and each of its values needs to be represented in 0 to 1 and we'll call that array 'x'.

*   We wil calculate X_max as max(X) and X_min as min(X) respectively and this is called 'fit'.

*   x = ( X - min(X) ) / ( max(X) - min(X) ).

*   Hence, x = [ 0, 1/5, 2/5, 3/5, 4/5, 1 ]

The min and max values of test data and train data can be different and can change the scale (different for both test data and train data) and due to this, the values become incomparable.

Hence, we keep the same fit for the test data as in train data and compare the values using the scale of train data.

As we use the same fit for test data, we need to apply "fit_transform" on train data that fits first by finding max and min and transforms to data between 0 and 1 and we apply just "transform" on test data keeping the same fit.

## Types of Data Analysis

1.   Univariate Analysis:-

      When we analyse the data of one column at a time.

      For Categorical Data Analysis:-

      -> sns.countplot() function is used to count the number of observations for each category of value of a column.

      -> Ex. for "Sex" column, sns.countplot(df,x='sex') will plot a histogram with X-axis as "Male" and "Female" and Y-axis with bar plots of the number of Males and Females respectively.

      -> "hue" parameter of .countplot() further divides the count based on another category.

      -> Ex. for "Sex" column, sns.countplot(df,x='sex',hue='survived') will plot a histogram with X-axis as "Male" and "Female" with two plots each having count of males survived and not survived and females who survived and not survived.
      

      For Numerical Data Analysis:-

      -> sns.histplot() function is used to plot a histogram.

      -> Ex. for "Age" column, sns.histplot(x=df['age'],bins=20) plots a histogram with bin size 20 on age column.

      -> You can also give parameter "kde=True" to also plot a KDE smooth curve on the histogram.

      -> sns.distplot() is used to plot a histogram and a KDE curve on top of it, just like sns.histplot with kde=True.

      -> sns.boxplot() is used to plot a whisker box plot of a column showing all the minimum value, maximum value, and all the quantiles, etc.

      -> Ex. for "Age" column, sns.boxplot(x=df['age']) plots the box plot.

2. Multivariate Analysis:-
  
    When we study multiple columns at once.

      For Numerical-Numerical Data Analysis:-

      -> Ex. sns.scatterplot(df,x='column1',y='column2')

      For Numerical-Categorical Data Analysis:-

      -> We use bar plot.

      -> sns.barplot(df['sex'],df['age'])
