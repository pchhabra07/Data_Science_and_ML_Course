## Recursion

> A function that calls itself



````python
def printN(n):
  if(n==0):
    return

  print(n)
  printN(n-1)

printN(5)
````

````python
def factorial(n):
  if(n==1):
    return 1
  else:
    return n*factorial(n-1)

print("Factorial of 5 is ",factorial(5))
````

## Decorators

> A function that takes a function as an argument and returns a function with some added functionality.





```
# @my_decorator
def func1():
```
This means func1=my_decorator(func1)

````python
def my_decorator(func):
  def wrapper():
    print("Before")
    func()
    print("Outer")
  return wrapper

@my_decorator
def func1():
  print("Hello!")

func1()
````
